```python
from bs4 import BeautifulSoup
import requests
import time, os
import re
```

```python

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

chrome_options = webdriver.ChromeOptions()
prefs = {"profile.managed_default_content_settings.images": 2}
chrome_options.add_experimental_option("prefs", prefs)
chrome_options.add_argument("--headless")

chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
query = "vegetarian"
Allrecipes_search = "https://www.allrecipes.com/search/results/?search="
Allrecipes_query = Allrecipes_search + query.replace(' ', '+')
```

```python
page = requests.get(Allrecipes_query).text
soup = BeautifulSoup(page, 'html5lib')
```

```python
Recipe_List = set()
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import WebDriverException
counter = 1
actions = ActionChains(driver)
for _ in range(45):
    driver = webdriver.Chrome(executable_path=chromedriver,chrome_options=chrome_options)
    try:
        driver.get(Allrecipes_query+'&page='+str(counter))
        counter += 1
        print(counter)
        time.sleep(5)

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        Recipe_class = soup.find('div', class_='search-results-content-results-wrapper grid-view')
        for link in Recipe_class.find_all('a', class_= 'card__titleLink manual-link-behavior' ):
            Recipe_List.add(link['href'])
    except Exception:
        print('failed request')
    finally:
        
        driver.quit()
    #for recipe_name in Recipe_class.find_all('h3', class_ = 'card__title'):
        #Recipe_List.add(recipe_name.text.strip())
    
```
