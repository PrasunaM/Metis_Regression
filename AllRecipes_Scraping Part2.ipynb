```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
import time, os
import re
```

```python
headers = ['Title','Rating', 'Prep_Time','Cook_Time','Yield', 'Nutrition', 'RatingCount', 'RecipeCategory']
```

```python
all_recipes_data = []
```

```python
import csv

Recipe_Links = []
with open('PaginatedRecipe.csv', newline='') as inputfile:
    for row in csv.reader(inputfile):
        Recipe_Links.append(row[0])

print(Recipe_Links)
```

```python
Recipe_Links[0]
```

```python
for url in Recipe_Links[0:100]:
    
    try:                        
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "lxml")
        title = soup.find('title').text.split('|')[0].rstrip() #Get title 
        
        rating= None
        if soup.find(class_ = 'component recipe-ratings'):
            if soup.find(class_ = 'component recipe-ratings').find ('span', class_ = 'review-star-text'):
                rating_text = (soup.find(class_ = 'component recipe-ratings').find ('span', class_ = 'review-star-text')).text
                if re.findall('\d+(?:\.\d+)?', rating_text ):
                    rating = re.findall('\d+(?:\.\d+)?', rating_text )[0]
                    
        
        ratings_count = None
        if soup.find(class_ = 'recipe-review-container euDisabled') and soup.find(class_ = 'recipe-review-container euDisabled').find('span', class_ = 'ugc-ratings-item'):
            ratings_count_text = (soup.find(class_ = 'recipe-review-container euDisabled').find('span', class_ = 'ugc-ratings-item').text)
            
            if ratings_count_text and re.findall('\d+(?:\.\d+)?', ratings_count_text ):
                ratings_count = re.findall('\d+(?:\.\d+)?', ratings_count_text )[0]
        
        
        recipe_category = None
        
        if soup.find(text="prep:"):
            prep_time = soup.find(text="prep:").findNext('div').text.replace('\n','').strip()
        else:
            prep_time = None
        
        if soup.find(text="cook:"):
            cook_time = soup.find(text="cook:").findNext('div').text.replace('\n','').strip()
        else:
            cook_time = None
            
        if soup.find(text="Yield:"):
            Yield = soup.find(text="Yield:").findNext('div').text.replace('\n','').strip()
        else:
            Yield = None
            
        if soup.find(text="Per Serving:"):
            nutrition = soup.find(text="Per Serving:").findNext('div').text.replace('\n','').strip()
        else:
            nutrition = None
        
        
        recipe_category = None
        breadcrumbs = soup.find(class_ = 'breadcrumbs__item--last')
        if breadcrumbs:
            recipe_category = breadcrumbs.find_previous_siblings('li')[0].find(class_ = 'breadcrumbs__title').text

        recipe_dict = dict(zip(headers, [title,
                                        rating,
                                        prep_time,
                                        cook_time,
                                        Yield,
                                        nutrition,
                                        ratings_count,
                                        recipe_category]))

        all_recipes_data.append(recipe_dict)
    except Exception:
        print(f'failed for {url}')
```
